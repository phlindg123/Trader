```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style("whitegrid")
```

```python
metrics = pd.read_csv("metrics.csv")
```

```python
btc = metrics[metrics.Symbol == "btc"].pivot(index="Time", columns = "Metric", values="Value")
btc.index = pd.to_datetime(btc.index)
btc = btc.last("2Y")
```

```python
eth = metrics[metrics.Symbol == "eth"].pivot(index="Time", columns = "Metric", values="Value")
eth.index = pd.to_datetime(eth.index)
eth = eth.last("2Y")
```

```python
facs = pd.DataFrame(index=btc.index)
facs["TAAR"] = btc.active_addresses.rolling("7D").mean() / btc.transactions_mean.rolling("7D").mean()
facs["Metcalfe"] = np.log(btc.mktcap) / np.log(btc.active_addresses.rolling("7D").mean())
facs["NVT"] = np.log(btc.mktcap) / np.log(btc.transfers_sum.rolling("7D").mean())
facs["MktCap"] = np.log(btc.mktcap)
```

```python
facs.diff().corr()
```

```python
facs[["Metcalfe", "MktCap"]].plot(secondary_y="MktCap")
```
