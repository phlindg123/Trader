```python
start = pd.to_datetime("2022-04-01")
end = pd.to_datetime("2022-08-15")
bars = b.get_bars("BTCUSDT", start, end, "1h").set_index("CloseTime")
```

```python
bars
```

```python
bars["Max"] = bars.Close.rolling(30).max()
bars["Min"] = bars.Close.rolling(30).min()
```

```python
bars[["Close", "Max", "Min"]].plot()
```

```python
max_waves = pd.concat([bars.Max.rename("waves"), pd.Series(1, index=bars.index)], axis=1).drop_duplicates("waves")
min_waves = pd.concat([bars.Min.rename("waves"), pd.Series(-1, index=bars.index)], axis=1).drop_duplicates("waves")
```

```python
waves = pd.concat([max_waves, min_waves]).sort_index()
waves = waves[waves[0] != waves[0].shift()].dropna()
```
