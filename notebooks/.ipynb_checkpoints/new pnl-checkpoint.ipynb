```python
class Data:
    def __init__(self, bars, window_size=30):
        self.bars = bars
        self.window_size = window_size
        
    def get(self, date, freq="1d"):
        prev_idx = date - pd.Timedelta(freq)*self.window_size
        df_ = self.bars[(self.bars.CloseTime > prev_idx) & (self.bars.CloseTime <= date)]
        return df_
    
```

```python
class Portfolio:
    def __init__(self, pairs, initial_balance = {"USDT" : 100.0}):
        self.pairs = pairs
        self.initial_balance = initial_balance
        self.setup()
        
    def setup(self):
        bases = list(set([x.base for x in self.pairs.values()]))
        quotes = list(set([x.quote for x in self.pairs.values()]))
        currs = bases + quotes
        self.holdings = {k:0 for k in currs}
        self.holdings.update(self.initial_balance)
        self.positions = {k : Position(k, 0, 1) for k in self.pairs.keys()}
    
    def order(self, symbol, fx, amount):
        pair = self.pairs[symbol]
        base_amt = amount
        quote_amt = amount * fx * (-1)
        if quote_amt < 0: #selling quote, buying base
            if self.holdings[pair.quote] < abs(quote_amt):
                print(f"Not enough {pair.quote} for {pair.symbol} order")
                return
        elif base_amt < 0:
            if self.holdings[pair.base] < abs(base_amt):
                print(f"Not enough {pair.base} for {pair.symbol} order")
                return
        
        self.holdings[pair.base] += base_amt
        self.holdings[pair.quote] += quote_amt
        self.positions[pair.symbol] = Position(symbol, base_amt=base_amt, fx=fx)
        
    def update(self, prices):
        #Bara base updateras hÃ¤r
        for symbol, pair in self.pairs.items():
            self.positions[symbol].update(prices[symbol])
    
    
    
    def get_holdings(self):
        return pd.Series(self.holdings)
    
    def get_positions(self):
        return pd.Series({k : v.base_amt for k, v in self.positions.items()})
    
    def has_position(self, symbol):
        return self.positions[symbol].base_amt != 0.0
    
    def get_position(self, symbol):
        return self.positions[symbol]
    
    def get_holding(self, curr):
        return self.holdings[curr]
```

```python
def cp_meanrev(price, window, sigma, trade_on=0):
    if price.shape[0] < window:
        return None
    mean = price.iloc[-window:].mean()
    std = price.diff().iloc[-window:].std()
    if (price.iloc[-1] < mean - std*sigma) and (trade_on == 0):
        return 1
    elif (price.iloc[-1] > mean) and (trade_on > 0):
        return 0
    else:
        return trade_on
```

```python
def cp_trend(price, window):
    if price.shape[0] < window*8:
        return None
    r1 = np.sign(price.iloc[-1]/price.iloc[-window*8] -1)
    r2 = np.sign(price.iloc[-1]/price.iloc[-window*4] -1)
    r3 = np.sign(price.iloc[-1]/price.iloc[-window] -1)
    signal = (r1+r2+r3)/3.0
    if signal > 0:
        return 1
    else:
        return 0
    return signal
```

```python
class Backtest:
    def __init__(self, pairs, data, index, state={}):
        self.pairs = pairs
        self.symbols = list(pairs.keys())
        self.data = data
        self.index = index
        self.state = state
        self.portfolio = Portfolio(pairs)
        self.states = []
        
        
    def initialize(self):
        pass
    
    def get_fx(self, symbol):
        #TODO: implement
        return 1
        
    def order(self, symbol, quote_size=0, base_size=0, fx=None):
        #TODO: ADD FEE
        base = symbol[:3] #NEED TO FIX, CAN BE MORE THAN 3.
        quote = symbol[3:] #NEED TO FIX
        if fx is None:
            fx = self.get_fx(symbol)
        if (quote_size == 0) & (base_size != 0):
            quote_size = base_size * fx
        elif (quote_size != 0) & (base_size == 0):
            base_size = quote_size / fx
        self.state["positions"][base] += base_size
        self.state["positions"][quote] -= quote_size
    
    def step(self, t):
        self.state["date"] = t
        bars = data.get(t)
        prices = bars.pivot(index="CloseTime", columns = "Symbol", values="Close")
        for symbol in self.symbols:
            price = prices[symbol]
            last_price = price.iloc[-1]
            signal = cp_trend(price[:-1], window=4)
            size = self.portfolio.get_holding("USDT") / len(self.symbols) / last_price
            if (signal == 1) and (not self.portfolio.has_position(symbol)):
                self.portfolio.order(symbol, last_price, size)
            elif (signal == 0) and self.portfolio.has_position(symbol):
                size = self.portfolio.get_position(symbol).base_amt
                self.portfolio.order(symbol, last_price, -size)
        self.portfolio.update(prices.iloc[-1])
        self.state["holdings"] = self.portfolio.get_holdings()
        self.state["positions"] = self.portfolio.get_positions()
        
            
    
    def run(self):
        for t in self.index:
            self.step(t)
            self.states.append(copy.deepcopy(self.state))
```
